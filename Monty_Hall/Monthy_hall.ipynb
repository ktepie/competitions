```python
import numpy as np
import matplotlib.pyplot as plt
my_game = np.random.randint(low=1, high=4, size=(10,2))
```

```python
def MontyHall():
    game = [1,2,3]
    choose = np.random.choice(game)
    car = np.random.choice(game)
    #print('you choose:' , choose ,'\n' 'car is in:' , car)
    #print('game is: ', game)
    
    if choose == car:
        game.remove(choose)
        phase2 = np.random.choice(game)
        #print('Monty opens door number:',phase2)
        game.remove(phase2)
        #print('the remaining door is:', game)
        
    else:
        game.remove(car)
        game.remove(choose)
        phase2 = game[0]
        #print('Monty opens door number', phase2)
        game = [car]
        #print('remaining door is:', game)
        
    #print('You decided to switch your door')
    if game == car:
        #print('YOU JUST WON')
        switch = 1
    else:
        #print('YOU JUST LOST')
        switch = 0
    
    #print('You decided to NOT to switch your door')
    if game == car:
        #print('YOU JUST LOST')
        not_switch = 0
    else:
        #print('YOU JUST WIN')
        not_switch = 1
        
    
    return switch,not_switch

wintp_switch = 0
wintp_nswitch = 0
arraysw =[]
xx = []
arraynsw = []

for x in range(1,100):
    [switchr,nswitchr]=MontyHall()
    xx.append(x)
    #print([switchr,nswitchr])
    #print('-------------------------')
    wintp_switch = wintp_switch + switchr
    #print('Your win\loss ratio by SWITCHING is:', wintp_switch,'-' ,x-wintp_switch)
    #print('Your current Win percentage by SWITCHING is:', (wintp_switch/x)*100)
    arraysw.append((wintp_switch/x)*100)
    #print('**************************')
    wintp_nswitch = wintp_nswitch + nswitchr
    #print('Your win\loss ratio by NOT SWITCHING is:', wintp_nswitch, '-', x-wintp_nswitch)
    #print('Your current Win percentage by NOT SWITCHING is:', (wintp_nswitch/x)*100)
    arraynsw.append((wintp_nswitch/x)*100)

fig= plt.figure(figsize=(20,10))  
plt.plot(xx,arraysw, label = 'Switching Doors')
plt.plot(xx,arraynsw, label = 'NOT Switching Doors')
plt.legend(loc =3, prop={'size':25})

plt.title('Monty Hall Game Show', fontsize = 35)
plt.xlabel('Number of tries', fontsize = 25)
plt.ylabel('Win percentage', fontsize = 25)
plt.grid(True)
plt.tight_layout()

    
```
